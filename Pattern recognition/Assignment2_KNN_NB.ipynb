## Method 01
- remove data samples with missing values, as done in the research paper
- train test division is 65.5% and 34.5%

```python
## Remove missing values
class_data = data.copy()
class_data = class_data[class_data['Bare Nuclei'] !='?']
```

```python
class_data
```

```python
class_data = class_data.astype(int)
```

```python
# Drop ID column
class_data = class_data.drop('Sample code number',axis =1)
```

```python
## Separate X and y
y = class_data.pop('Class')
X = class_data
```

```python
## Train test Split  Test size = 0.345 train soze = 0.655
X_train_method1, X_test_method1, y_train_method1, y_test_method1 = train_test_split(X, y, test_size=0.345, random_state=42)
```

```python
# K-NN classifier to the training set  
## Euclidean distance is used (p=2)
## For low-dimensional data, the KD Tree Algorithm might be the best solution
## For a high-dimensional space, the Ball Tree Algorithm might be the best solution.
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='ball_tree', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None) 
classifier.fit(X_train_method1, y_train_method1)  

#Predicting the test set result  
y_pred_method1= classifier.predict(X_test_method1) 

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test_method1,y_pred_method1))
```
