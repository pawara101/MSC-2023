## Method 01
- remove data samples with missing values, as done in the research paper
- train test division is 65.5% and 34.5%

```python
## Remove missing values
class_data = data.copy()
class_data = class_data[class_data['Bare Nuclei'] !='?']
```

```python
class_data
```

```python
# Drop ID column
class_data = class_data.drop('Sample code number',axis =1)
```

```python
## Separate X and y
y = class_data.pop('Class')
X = class_data
```

```python
## Train test Split  Test size = 0.345 train soze = 0.655
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.345, random_state=42)
```

```python
from sklearn import linear_model
logr = linear_model.LogisticRegression()
## Fit to the model
logr.fit(X_train,y_train)
```

```python
## Prediction using test data
y_predicted = logr.predict(X_test)
```

```python
## Calculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predicted))
```
